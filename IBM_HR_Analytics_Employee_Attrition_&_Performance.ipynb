```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate attrition proportions
attrition = df['Attrition'].value_counts(normalize=True)

# Create the bar plot
plt.figure(figsize=(8,6))
ax = sns.barplot(x=attrition.index, y=attrition, palette="coolwarm")

# Annotate bars with percentages
for p in ax.patches:
    ax.annotate(f'{p.get_height() * 100:.2f}%',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

# Set labels and title
plt.title('Distribution of Attrition Rate')
plt.xlabel('Attrition')
plt.ylabel('Percentage')

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

def calculate_attrition_rate(df, column):
    """
    Calculate the attrition rate for each category in the specified column.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the data.
    column (str): The column name to group by.

    Returns:
    pd.DataFrame: A DataFrame showing attrition rates.
    """
    # Count occurrences of each category split by Attrition
    attrition_counts = df.groupby([column, 'Attrition']).size().unstack(fill_value=0)

    # Calculate the attrition rate (percentage of 'Yes' in each category)
    attrition_rate = (attrition_counts['Yes'] / attrition_counts.sum(axis=1)) * 100

    # Convert to DataFrame
    attrition_rate_df = attrition_rate.reset_index()
    attrition_rate_df.columns = [column, 'AttritionRate (%)']

    return attrition_rate_df

# Example usage
attrition_rate_df = calculate_attrition_rate(df, 'Department')  # Replace 'Department' with any column
print(attrition_rate_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Plot 1: KDE plot of Age with Attrition hue
sns.kdeplot(data=df, x='Age', hue='Attrition', fill=True, ax=axes[0], palette="coolwarm")
axes[0].set_title('Attrition by Age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Density')

# Plot 2: Bar plot of Gender count with Attrition hue
attrition_rate_df = calculate_attrition_rate(df, 'Gender')
sns.barplot(data=attrition_rate_df, x='Gender', y='AttritionRate (%)', ax=axes[1], palette="coolwarm")
axes[1].set_title('Attrition Rate by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Attrition Rate (%)')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
